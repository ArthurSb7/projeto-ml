```python
%%writefile app.py
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression

# 1. Carregar os dados (certifique-se de ter o pizzas.csv na mesma pasta do Jupyter)
df = pd.read_csv('pizzas.csv')

# 2. Preparar os dados e Treinar o Modelo
X = df[['diametro']]
y = df['preco']

modelo = LinearRegression()
modelo.fit(X, y)

# 3. Construir a Interface com Streamlit
st.title("Prevendo o valor de uma pizza рџЌ•")
st.divider()

# Campo para o usuГЎrio digitar o diГўmetro
diametro = st.number_input("Digite o tamanho do diГўmetro da pizza (em cm):", min_value=0.0, format="%.2f")

# SГі faz a previsГЈo se o valor digitado for maior que zero
if diametro > 0:
    # O modelo espera uma lista de listas para prever
    preco_previsto = modelo.predict([[diametro]])
    
    # Exibir o resultado formatado
    st.write(f"O valor da pizza com diГўmetro de **{diametro} cm** Г© de **R$ {preco_previsto[0]:.2f}**")
    
    # Efeito visual de balГµes (como no vГ­deo)
    st.balloons()
```
